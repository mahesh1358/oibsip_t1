#### Step 4: Data Visualization

```python
sns.heatmap(data.corr(),annot=True,cmap="coolwarm")
```

From above graph we can observe that, petal width and petal length have high correlation
petal length and sepal width have good correlations
petal width and sepal length also have good correlations.

```python
sns.scatterplot(data=data,x='SepalLengthCm',y='SepalWidthCm',hue='species')
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

In the above plot we will see the relation between sepal length and sepal width.
From this we can interpret that the species setosa has smallefr sepal lengths but larger sepal widths,
versicolor species lies in the middle of the other two species in terms of sepal length and width and 
species virginica has larger sepal lengths but smaller sepal widths. 

```python
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm",hue="species",data=data)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

From the above plot we will see the relation between petal length and petal width.

from the graph we will observe that species setosa has smaller petal lengths and widths.
Versicolor species lies in the middle of the other two species in terms of petal length and width.
Species verginica has the largest petal length and widths.
